```python
%matplotlib inline
from nuscenes.nuscenes import NuScenes
```

```python
nusc = NuScenes(version='v1.0-mini', dataroot='/Users/curt/projects/avs/nuscenes-data/v1.0-mini', verbose=True)
```

```python
nusc.list_scenes()
```

```python
scene = nusc.scene[1]
scene
```

```python
from pyproj import Proj, transform
inProj = Proj(init='epsg:5181')
outProj = Proj(init='epsg:4326')
```

```python
sample = nusc.get('sample', scene['first_sample_token'])
data = {}
for i in range(10):
    for sensor in ['CAM_FRONT', 'CAM_FRONT_LEFT', 'CAM_FRONT_RIGHT', 'CAM_BACK', 'CAM_BACK_LEFT', 'CAM_BACK_RIGHT']:
        sample_data = nusc.get('sample_data', sample['data'][sensor])
        ego_pose = nusc.get('ego_pose', sample_data['ego_pose_token'])
        x2,y2 = transform(inProj,outProj,ego_pose['translation'][0]+200000, ego_pose['translation'][1]+500000)
        filename = sample_data['filename'].split('/')[2]
        data[filename] = {
            "latitude": y2,
            "longitude": x2,
            "altitude": 20.0,
            "dop": 5.0
        }
    #print(sample_data['filename'], x2, y2)
    sample = nusc.get('sample', sample['next'])

import json
print(json.dumps(data, indent=4))
```
